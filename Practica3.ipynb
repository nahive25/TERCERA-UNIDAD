```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parmetros de la distribucin lognormal
mu = 1.0   # Media de la variable logartmica
sigma = 0.5  # Desviacin estndar de la variable logartmica
n = 100  # Nmero de variables aleatorias a generar

# Generar n nmeros aleatorios U ~ Uniform(0, 1)
U = np.random.uniform(0, 1, n)

# Transformada inversa para generar tiempos de inactividad lognormales
Z = mu + sigma * stats.norm.ppf(U)
T = np.exp(Z)  # Tiempos generados

# Mostrar los resultados
print("Tiempos de inactividad generados:", T)

# Graficar histograma de los tiempos de inactividad generados
plt.hist(T, bins=20, edgecolor='black', density=True)

# Graficar la funcin terica de densidad de la distribucin lognormal
x = np.linspace(min(T), max(T), 100)
pdf = stats.lognorm.pdf(x, sigma, scale=np.exp(mu))
plt.plot(x, pdf, 'r', label='Distribucin lognormal terica')

plt.title("Histograma de los tiempos de inactividad (Distribucin Lognormal)")
plt.xlabel("Tiempo de inactividad")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.show()
```
