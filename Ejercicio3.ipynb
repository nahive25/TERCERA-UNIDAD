```python
import numpy as np
import matplotlib.pyplot as plt

media_espera = 10
desviacion_espera = 2
num_clientes = 50

mu = np.log(media_espera) - 0.5 * np.log(1 + (desviacion_espera / media_espera) ** 2)
sigma = np.sqrt(np.log(1 + (desviacion_espera / media_espera) ** 2))

tiempos_espera = np.random.lognormal(mean=mu, sigma=sigma, size=num_clientes)

print("Tiempos de espera (minutos):", tiempos_espera)

plt.hist(tiempos_espera, bins=15, color='royalblue', edgecolor='black', density=True)
plt.xlabel("Tiempo de espera (minutos)")
plt.ylabel("Densidad")
plt.title("Distribución de tiempos de espera en servicio técnico")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

La simulación de los tiempos de espera mediante la distribución lognormal permite modelar con mayor precisión la variabilidad en los tiempos de atención en el mostrador de servicio técnico. Este análisis ayuda a la tienda a prever los momentos de mayor congestión y tomar decisiones estratégicas para mejorar la gestión de la fila de espera.

Al conocer la distribución de los tiempos de espera, se pueden optimizar los recursos, reducir la congestión en horas pico y mejorar la experiencia del cliente, minimizando tiempos de espera excesivos y asegurando un servicio más eficiente.
