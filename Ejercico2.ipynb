```python
import numpy as np
import matplotlib.pyplot as plt

tasa_llegadas = 1 / 15
horas_operacion = 12
minutos_operacion = horas_operacion * 60
num_llegadas =  np.random.poisson(tasa_llegadas * minutos_operacion)

tiempos_entre_llegadas = np.random.exponential(scale=1/tasa_llegadas, size=num_llegadas)

print("Tiempos entre llegadas (minutos):", tiempos_entre_llegadas)

plt.hist(tiempos_entre_llegadas, bins=20, color='royalblue', edgecolor='black', density=True)
plt.xlabel("Tiempo entre llegadas (minutos)")
plt.ylabel("Densidad")
plt.title("Distribución de tiempos entre llegadas")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

La simulación de los tiempos entre llegadas de vehículos eléctricos mediante la distribución exponencial permite modelar de manera realista la aleatoriedad en la demanda de carga. Al conocer estos tiempos, la estación puede prever la cantidad de puntos de carga necesarios para minimizar tiempos de espera y optimizar el flujo de atención.

Este análisis facilita una mejor planificación de los recursos, permitiendo ajustar la capacidad operativa en función de la demanda esperada. Además, al identificar patrones en la llegada de vehículos, se pueden tomar decisiones estratégicas para mejorar la eficiencia del servicio y garantizar una experiencia óptima para los usuarios.
