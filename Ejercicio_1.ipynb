```python
import numpy as np
import matplotlib.pyplot as plt

n_pedidos = 20
p_exito = 0.7
dias = 7

pedidos_satisfechos = np.random.binomial(n_pedidos, p_exito, dias)

print("Pedidos satisfechos por día:", pedidos_satisfechos)

plt.bar(range(1, dias + 1), pedidos_satisfechos, color='royalblue', edgecolor='black')
plt.xticks(range(1, dias + 1))
plt.xlabel("Día")
plt.ylabel("Pedidos Satisfechos")
plt.title("Simulación de Pedidos Satisfechos por Día")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Las variables aleatorias permiten modelar de manera realista la incertidumbre y variabilidad en los tiempos de servicio, reflejando mejor el comportamiento de los sistemas. Su uso facilita la optimización de recursos, identificando cuellos de botella y permitiendo una asignación más eficiente para reducir tiempos de espera y mejorar la calidad del servicio. Además, contribuyen a la reducción de costos al prever posibles problemas y ajustar la capacidad operativa de manera óptima.

La simulación basada en variables aleatorias permite evaluar diferentes escenarios sin afectar el sistema real, lo que facilita la toma de decisiones estratégicas antes de su implementación. También mejora la planificación operativa al proporcionar información clave para optimizar el flujo de atención y mejorar la experiencia del cliente. En general, su aplicación es fundamental en sectores como logística, salud, manufactura y atención al cliente, donde la eficiencia y la predicción de tiempos de servicio son esenciales para el éxito operativo.
