```python
import numpy as np
import matplotlib.pyplot as plt

n_pedidos = 20
p_exito = 0.7
dias = 7

pedidos_satisfechos = np.random.binomial(n_pedidos, p_exito, dias)

print("Pedidos satisfechos por d√≠a:", pedidos_satisfechos)

plt.bar(range(1, dias + 1), pedidos_satisfechos, color='royalblue', edgecolor='black')
plt.xticks(range(1, dias + 1))
plt.xlabel("D√≠a")
plt.ylabel("Pedidos Satisfechos")
plt.title("Simulaci√≥n de Pedidos Satisfechos por D√≠a")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Las variables aleatorias permiten modelar de manera realista la incertidumbre y variabilidad en los tiempos de servicio, reflejando mejor el comportamiento de los sistemas. Su uso facilita la optimizaci√≥n de recursos, identificando cuellos de botella y permitiendo una asignaci√≥n m√°s eficiente para reducir tiempos de espera y mejorar la calidad del servicio. Adem√°s, contribuyen a la reducci√≥n de costos al prever posibles problemas y ajustar la capacidad operativa de manera √≥ptima.

La simulaci√≥n basada en variables aleatorias permite evaluar diferentes escenarios sin afectar el sistema real, lo que facilita la toma de decisiones estrat√©gicas antes de su implementaci√≥n. Tambi√©n mejora la planificaci√≥n operativa al proporcionar informaci√≥n clave para optimizar el flujo de atenci√≥n y mejorar la experiencia del cliente. En general, su aplicaci√≥n es fundamental en sectores como log√≠stica, salud, manufactura y atenci√≥n al cliente, donde la eficiencia y la predicci√≥n de tiempos de servicio son esenciales para el √©xito operativo.
