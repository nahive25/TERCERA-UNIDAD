```python
import numpy as np
import matplotlib.pyplot as plt
import math

lambda_poisson = 5
n = 100

U = np.random.uniform(0, 1, n)

def poisson_inverse_transform(lambda_poisson, u):
    k = 0
    F_k = math.exp(-lambda_poisson)
    while u > F_k:
        k += 1
        F_k += (lambda_poisson**k * math.exp(-lambda_poisson)) / math.factorial(k)
    return k

poisson_vectorizado = np.vectorize(poisson_inverse_transform)
demanda_simulada = poisson_vectorizado(lambda_poisson, U)

print("Demanda diaria simulada:", demanda_simulada)

plt.hist(demanda_simulada, bins=np.arange(0, max(demanda_simulada) + 2) - 0.5,
         edgecolor='black', density=True)
plt.title("Simulación de demanda diaria (Distribución Poisson)")
plt.xlabel("Número de productos demandados")
plt.ylabel("Frecuencia")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
